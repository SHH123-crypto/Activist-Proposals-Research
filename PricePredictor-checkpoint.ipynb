```python
import csv
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
from datetime import datetime
dates = []
prices = []

def get_data(filename):
    global dates, prices
    with open(filename, newline='') as csvfile:
        c = csv.reader(csvfile)
        next(c)  # Skip header
        counter = 0
        for row in c:
            try:
                prices.append(float(row[1]))
                # Parse day from date, assuming format like MM/DD/YYYY
                counter = counter + 1
                dates.append(counter)
            except Exception as e:
                print(f"Error parsing row {row}: {e}")
    return

def predict_prices(dates, prices, x):
    dates = np.reshape(dates, (-1, 1))
    x = np.array([[x]])  # Ensure x is 2D

    svr_len = SVR(kernel='linear', C=1e3)
    svr_poly = SVR(kernel='poly', C=1e3, degree=2)
    svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)

    svr_rbf.fit(dates, prices)
    svr_poly.fit(dates, prices)
    svr_len.fit(dates, prices)

    # Plotting
    plt.scatter(dates, prices, color='black', label='Data')
    plt.plot(dates, svr_rbf.predict(dates), color='red', label='RBF Model')
    plt.plot(dates, svr_poly.predict(dates), color='green', label='Poly Model')
    plt.plot(dates, svr_len.predict(dates), color='blue', label='Lin Model')
    plt.xlabel('Days since Start')
    plt.ylabel('Price')
    plt.title('Support Vector Regression')
    plt.legend()
    plt.show()

    return svr_rbf.predict(x)[0], svr_len.predict(x)[0], svr_poly.predict(x)[0]

# Run
date_prediction = 255
get_data('AAPL Price data - Sheet1.csv')
predicted_price = predict_prices(dates, prices, date_prediction)

print(f"Predicted price on day {date_prediction}: RBF={predicted_price[0]}, Linear={predicted_price[1]}, Poly={predicted_price[2]}")
```
